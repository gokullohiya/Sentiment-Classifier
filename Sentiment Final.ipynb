Code for Finding the given review is a Positive or Negative.

```python
punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@','[',']']
def strip_punctuation(string):
    for char in punctuation_chars:
        string = string.replace(char,'')
    return string
with open('E:\Sentiment\opinion-lexicon-English\positive-words.txt','r') as ps:
    #print(ps.read())
    positive_words = []
    for lin in ps:
        if lin[0] != ';' and lin[0] != '\n':
            positive_words.append(lin.strip())
with open('E:\Sentiment\opinion-lexicon-English\\negative-words.txt','r') as ns:
    negative_words = []
    for lin in ns:
        if lin[0] != ';' and lin[0] != '\n':
            negative_words.append(lin.strip())
def get_pos(s):
    s = strip_punctuation(s)
    lst = s.split()
    for word in lst:
        if word.lower() in positive_words:
            return 'Positive Sentiment'
        if word.lower() in negative_words:
            return 'Negative Sentiment'
```

You can test it with your own piece of code.

```python
readingFile = open('E:\Sentiment\CustomerReviews -3 domains (IJCAI2015)\Computer.txt','r')
lines = readingFile.readlines()
for line in lines:
    print(get_pos(line))
```
